```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
N = 810     #объем аудитории
N0 = 11     #количество людей, знающих о товаре в начальный момент времени
t = np.arange(0, 30, 0.1)     #временной промежуток(длительность рекламной кампании)
```

## Первый случай

```python
alpha1 = 0.64     #интенсивность рекламной кампании
alpha2 = 0.00014  #сарафанное радио

t = np.arange(0, 800, 0.1)     #временной промежуток(длительность рекламной кампании)

#уравнение, описывающее распространение рекламы
def xd(t, x):
    return ( alpha1 + alpha2*x )*( N - x )

x = odeint(xd, N0, t)

plt.plot(t,  x)
```

## Второй случай

```python
alpha1 = 0.000014     #интенсивность рекламной кампании
alpha2 = 0.63  #сарафанное радио

t = np.arange(0, 800, 0.1)     #временной промежуток(длительность рекламной кампании)

#уравнение, описывающее распространение рекламы
def xd(t, x):
    return ( alpha1 + alpha2*x )*( N - x )

x = odeint(xd, N0, t)

plt.plot(t,  x)
```

## Третий случай

```python
N = 200

#интенсивность рекламной кампании
def alpha1(t):
    return 0.7*t

#сарафанное радио
def alpha2(t):
    return 0.4*np.cos(t)

t = np.arange(0, 30, 0.01)     #временной промежуток(длительность рекламной кампании)

#уравнение, описывающее распространение рекламы
def xd(t, x):
    return ( alpha1(t) + alpha2(t)*x )*( N - x )

x = odeint(xd, N0, t)


plt.figure(figsize=(8, 8))
plt.plot(t,  x)
plt.grid()
#plt.yticks(np.arange(0, N, step=10))
```
