```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pylab
from matplotlib import mlab
```

```python
a = 0.19     #коэффициент естественной смертности хищников
b = 0.048    #коэффициент естественного прироста жертв   
c = 0.39     #коэффициент прироста жертв
d = 0.036    #коэффициент смертности жертв
```

```python
def syst(t, z):
    return [-a*z[0] + b*z[0]*z[1], c*z[1] - d*z[0]*z[1]]
```

```python
v0 = [13, 18]
t0 = 0
tmax = 300
dt = 0.01
t = np.arange(t0, tmax, dt)
tborders = [t0, tmax]
```

```python
y = solve_ivp(syst, tborders, v0, t_eval=t)
```

```python
plt.figure(figsize=(13, 13))
plt.plot(y.t, y.y.T)
plt.grid()
plt.xlabel('время', fontsize=18)
plt.ylabel('численность популяций', fontsize=18)
plt.legend(['хищники', 'жертвы'], shadow=True, fontsize=20)
plt.title('Изменение численности хищников и численности жертв', fontsize=15)
pylab.xlim (0, 100)
```

```python
plt.figure(figsize=(13, 13))
plt.plot(y.y[0], y.y[1])
plt.grid()
plt.xlabel('хищники', fontsize=18)
plt.ylabel('жертвы', fontsize=18)
plt.title('Зависимость численности хищников от численности жертв', fontsize=15)
```

```python
v0 = [13,18]   #вектор начальных условий
t0 = 0
tmax = 30
dt = 0.01
t = np.arange(t0, tmax, dt)
tborders = [t0, tmax]
y = solve_ivp(syst, tborders, v0, t_eval=t)   #решение системы на промежутке t
plt.figure(figsize=(13, 13))
plt.plot(y.y[0], y.y[1])
plt.grid()
plt.xlabel('хищники', fontsize=18)
plt.ylabel('жертвы', fontsize=18)
plt.title('Зависимость численности хищников от численности жертв', fontsize=15)
```
