```python
import numpy as np
from scipy. integrate import odeint
import matplotlib.pyplot as plt
import math
```

```python
w2 = 1.5
tmax = 60
step = 0.05
y0 = [0, 0]
```

```python
def W(y, t):
    y1, y2 = y
    return [y2, -w2*y1 ]
```

```python
t = np.arange( 0, tmax, step)
w1 = odeint(W, y0, t)
y11 = w1[:,0]
y21 = w1[:,1]
```

```python
fig = plt.figure(facecolor='white')
plt.plot(t, y11, linewidth=2)
plt.ylabel("x")
plt.xlabel("t")
plt.grid(True)
plt.show()
```

```python
fig2 = plt.figure(facecolor='white')
plt.plot(y11, y21, linewidth=2)
plt.ylabel("y")
plt.xlabel("x")
plt.grid(True)
plt.show()
```

```python
## 3
w2 = 11
g = 1
```

```python
def f(t): 
    f = 2*math.cos(t)
    return f
```

```python
def W(y, t):
    y1, y2 = y
    return [y2, -w2*y1 - g*y2 + f(t) ]
```

```python
t = np.arange( 0, tmax, step)
w1 = odeint(W, y0, t)
y11 = w1[:,0]
y21 = w1[:,1]
```

```python
fig = plt.figure(facecolor='white')
plt.plot(t, y11, linewidth=2)
plt.ylabel("x")
plt.xlabel("t")
plt.grid(True)
plt.show()
```

```python
fig2 = plt.figure(facecolor='white')
plt.plot(y11, y21, linewidth=2)
plt.ylabel("y")
plt.xlabel("x")
plt.grid(True)
plt.show()
```
